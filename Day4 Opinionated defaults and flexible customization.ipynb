# Opinionated defaults and flexible customization

Seaborn은 단일 함수 호출로 완벽한 그래픽을 만들 수 있다
=>  함수가 자동으로 informative 한 축이름 과 범례를 추가하게 할 수 있다 
    축이름과 범례는 플롯내의 semantic mappings를 설명 .  
semantic mapping : 직역하면 의미론적인 맵핑


Seaborn은 데이터의 성격에 부합하는(ex 범주형, 수치형 등등...) 디폴트값의 파라미터 사용
예를 들어서 컬러 맵핑을 할 때,
범주형 변수의 다른 수준을 나타내기 위해 고유한 색상(파란색, 주황색, 때로는 녹색)을 사용함
수치형 변수를 맵핑할때 어떤 기능은 연속 그라데이션으로 전환

```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g"
)
```

<!-- When you're ready to share or publish your work, you’ll probably want to polish the figure beyond what the defaults achieve. Seaborn allows for several levels of customization. It defines multiple built-in themes that apply to all figures, its functions have standardized parameters that can modify the semantic mappings for each plot, and additional keyword arguments are passed down to the underlying matplotlib artists, allowing even more control. Once you’ve created a plot, its properties can be modified through both the seaborn API and by dropping down to the matplotlib layer for fine-grained tweaking(=modify): -->

```python
결과물 공유할 때, 기본값이 낼 수있는 성과보다 더 멋지게 수치를 좀 다듬고 싶을 것이다.
(기본적으로 보여주는 것보다 좀 더 이쁘게 보여줄 수 있다는 뜻인것 같다)
Seaborn이 몇가지 레벨의 커스터마이징 보유
모든 그림에 적용가능하고, 각 플롯의 semantic mapping(의미 체계 매핑)을 수정할 수 있고 
표준화된 파라미터 가지고 있음

플롯을 만든 후에는 seaborn API를 통해 속성을 수정하고 
세분화 된 조정을 위해 matplotlib 레이어로 드롭다운하여 수정가능
```

```python
# import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",
    palette="crest", marker="x", s=100,
)
g.set_axis_labels("Bill length (mm)", "Bill depth (mm)", labelpad=10)
g.legend.set_title("Body mass (g)")
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15)
g.despine(trim=True)
plt.figure()
```
